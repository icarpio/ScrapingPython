# Leer una imagen

```python
import numpy as np
import cv2
from ipywidgets import widgets
```

```python
imagen = cv2.imread('prueba.jpg', 0)
```

```python
cv2.imshow('Icarpio OpenCV', imagen)
```

```python
# cv2.IMREAD_COLOR para color, las transparencias se ignoran 1
# cv2.IMREAD_GRAYSCALE imagen en escala de grises 0 
# cv2.IMREAD_UNCHANGED  carga la imagen sin alteraciones -1
```

# Mostrar la imagen

```python
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Guardar la imagen

```python
cv2.imwrite('pruebaGris.png', imagen)
```

# Usando Matplotlib 

```python
#OPENCV is BGR
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('prueba.jpg', -1)
```

```python
plt.imshow(img, cmap= 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Matplotlib is RGB
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('prueba.jpg')
imgplot = plt.imshow(img,  cmap= 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```
